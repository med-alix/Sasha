```python
# Profile a parameter
from sasha.wrappers.run_inference import ShallowSAInference
sa_infer = ShallowSAInference(config_path='../config_model.yml',
                             **{
                                "prop_type": 'rrsp',
                                "sensor"   : 'sentinel_2A', # select from ['sentinel_2A', 'sentinel_2B', 'dirac', 'hyspex_muosli']
                                "backend"  : 'JAX', # select from ['JAX', 'NUMPY', 'TENSORFLOW']
                                # 'logging': logging.DEBUG,
                                })

# Generate a sample and simulate reflectance
params        = {"z":1, "cdom": -6, "chl":-3, "tsm": -4, "alpha_m_1": .2} # True values, clear water
sample        = sa_infer.generate_sample(params=params, n_samples=1)[0]

# Profile a parameter
paramater_key = "z" # select either "z",  "tsm", "cdom" , "chl" or "alpha_m_1" to profile other parameters
profile_data  = sa_infer.profile_parameter(paramater_key, sample,
                                            init_param = params, # Initialize from true value for simulation purposes
                                            num_points=25,
                                            values ='diffuse')
                                            
sa_infer.plot_profile(profile_data)                       
```
